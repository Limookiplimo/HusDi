# Hustler Distributors
## Data Analysis

```python
# Import Library
import pandas as pd
```

Sales Data

```python
df = pd.read_csv("Hustler Distributors - Sales.csv")
```

```python
df.head()
```

Accounts Data

```python
acc_df = pd.read_csv("Hustler Distributors - Accounts.csv")
```

```python
acc_df.head()
```

```python
# Convert to date types
acc_df["EXPECTED DATE"] = pd.to_datetime(acc_df["EXPECTED DATE"])
acc_df["PAYMENT DATE"] = pd.to_datetime(acc_df["PAYMENT DATE"])
```

```python
# Get delay count
acc_df["DELAY"] = (acc_df["PAYMENT DATE"] - acc_df["EXPECTED DATE"]).dt.days

# Remove negative delays
acc_df['DELAY'] = acc_df['DELAY'].apply(lambda x: 0 if x < 0 else x)
```

```python
# Pending balance
acc_df["PENDING BALANCE"] = (acc_df["INVOICE AMOUNT"] - acc_df["AMOUNT PAID"])
```

```python
# Paid on time
acc_df["TIMELY PAYMENT"] = acc_df.apply(lambda row: "Yes" if row["DELAY"] == 0 and row["PENDING BALANCE"] == 0 else "No", axis=1)
```

```python
acc_df.head()
```

Logistics Data

```python
log_df = pd.read_csv("Hustler Distributors - Logistics.csv")
```

```python
log_df.head()
```

```python
# Change data types
log_df["EXPECTED DATE"] = pd.to_datetime(log_df["EXPECTED DATE"])
log_df["DELIVERY DATE"] = pd.to_datetime(log_df["DELIVERY DATE"])
```

```python
# Get delay count
log_df["DELAY"] = (log_df["DELIVERY DATE"] - log_df["EXPECTED DATE"]).dt.days
log_df['DELAY'] = log_df['DELAY'].apply(lambda x: 0 if x < 0 else x)
```

```python
# Customer routes
Donholm = ["Ahmed shop"]
Kitengela = ["Sarah shop"]
Tala = ["John shop","Jane shop"]
Eastleigh = ["Willy shop","Emily shop","Alhamdu shop"]
Umoja = ["Tom shop","Grace shop","Green shop"]
```

```python
# Regex to adjoin routes
Donholm = '|'.join(Donholm)
Kitengela = '|'.join(Kitengela)
Tala = '|'.join(Tala)
Eastleigh = '|'.join(Eastleigh)
Umoja = '|'.join(Umoja)
```

```python
# Assign each customer a route
log_df["ROUTE"][log_df["ACUSTOMER NAME"].str.contains(Donholm)] = "Donholm"
log_df["ROUTE"][log_df["ACUSTOMER NAME"].str.contains(Kitengela)] = "Kitengela"
log_df["ROUTE"][log_df["ACUSTOMER NAME"].str.contains(Tala)] = "Tala"
log_df["ROUTE"][log_df["ACUSTOMER NAME"].str.contains(Eastleigh)] = "Eastleigh"
log_df["ROUTE"][log_df["ACUSTOMER NAME"].str.contains(Umoja)] = "Umoja"
```

```python
# Distance
distance_dict = {"Ahmed shop": 20, "John shop": 35, "Jane shop": 37,"Willy shop": 15, "Emily shop": 17,"Sarah shop": 30,"Alhamdu shop": 15,"Tom shop": 20,"Grace shop": 19,"Green shop": 6,}
```

```python
log_df["DISTANCE"] = log_df["ACUSTOMER NAME"].map(distance_dict)
```

```python
# Save files 
df.to_csv("sales.csv")
acc_df.to_csv("accounts.csv")
log_df.to_csv("logistics.csv")
```
